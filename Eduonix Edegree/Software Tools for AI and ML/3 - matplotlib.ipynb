```python
### 3.4

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('BreadBasket_DMS.csv')
df = df.groupby(['Date']).agg(['count'])

x = range(len(df.index))

yVals = []
[yVals.append(row[0]) for ind, row in df["Transaction"].iterrows()]
  
fit = np.polyfit(x,yVals,3)
fit_fn = np.poly1d(fit)

plt.subplots()
plt.plot(x,yVals, '-k', x, fit_fn(x), '--r')
# plt.locator_params(axis='x', nbins=20)
plt.ylabel('Total Purchases')
plt.xlabel('Day')
plt.grid()
plt.show()
```

```python
## 3.0 Problem/Answer section

import pandas as pd
from matplotlib import pyplot as plt
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df = iris[["petal_length","sepal_length","species"]]
# y = iris["species"]

# df = pd.concat([X,y],axis=1)

fig = plt.figure()
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Petal Length', fontsize = 15)
ax.set_ylabel('Sepal Length', fontsize = 15)
ax.set_title('Iris Comparison', fontsize = 20)
species = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']
for target, color in zip(species,colors):
    indices = df['species'] == target
    ax.scatter(df.loc[indices, 'petal_length'], df.loc[indices, 'sepal_length'], c = color, s = 50)
ax.legend(species)
ax.grid()
```
