- load .jpg of your chosing using the cv2package .imread() command to load image as object
    - import cv2
    - x=cv2.imread('file')
- scale the data to range from 0-1 by dividing all values by 255
    - hint: use.shape to determine dimensions of the object
- extract the RGB of image as separate objects (red,green,blue)
- plot the images using plt.imshow()

```python
from matplotlib import pyplot as plt
import cv2

image_data=cv2.imread('test.jpg')
scaled_image_data = image_data / 255.
scaled_image_data.shape

image_slice_red =  scaled_image_data[:,:,0]
image_slice_green =  scaled_image_data[:,:,1]
image_slice_blue =  scaled_image_data[:,:,2]

plt.subplot(221)
plt.imshow(image_slice_red, cmap=plt.cm.Reds_r)
plt.subplot(222)
plt.imshow(image_slice_green, cmap=plt.cm.Greens_r)
plt.subplot(223)
plt.imshow(image_slice_blue, cmap=plt.cm.Blues_r)  
plt.subplot(224)
plt.imshow(image_data)  
plt.show()
```
