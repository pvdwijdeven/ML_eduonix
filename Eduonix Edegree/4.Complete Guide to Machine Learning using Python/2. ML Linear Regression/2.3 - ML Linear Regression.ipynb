```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from scipy import stats

diabetes = datasets.load_diabetes() # Call the diabetes dataset from sklearn

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names) # load the dataset as a pandas data frame


### One Feature
features=["bmi"]

x = df[features]
y = diabetes.target # define the target variable (dependent variable) as y


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

# print(x_train)

lm = linear_model.LinearRegression()


model = lm.fit(x_train,y_train)
predictions = lm.predict(x_test)

fit = np.polyfit(x["bmi"],y,1)
fit_fn = np.poly1d(fit)
print(fit_fn)


fig, ax = plt.subplots()
plt.plot(x["bmi"],y,'ro')
plt.plot(x["bmi"], fit_fn(x["bmi"]), '--k')
plt.plot(x_test,predictions,'bo')
plt.xlabel('BMI')
plt.ylabel('Score')
ax.grid()
plt.show()


slope, intercept, r_value, p_value, std_err = stats.linregress(x["bmi"],y)
# print("The r-value is: ",r_value)
# print("The p-value is: ",p_value)
print(std_err)

pred = lm.predict(x)
sq = np.sum((y - list(pred)) ** 2)

print(np.sqrt(sq / len(x)))
print(lm.score(x,y))
```

```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split

diabetes = datasets.load_diabetes() # Call the diabetes dataset from sklearn

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names) # load the dataset as a pandas data frame



### Multiple Features
features=list(df)

x = df[features]
y = diabetes.target # define the target variable (dependent variable) as y

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
lm = linear_model.LinearRegression()


model = lm.fit(x_train,y_train)
predictions = lm.predict(x_test)

fit = np.polyfit(x["bmi"],y,1)
fit_fn = np.poly1d(fit)
print(fit_fn)


fig, ax = plt.subplots()
plt.plot(x_train, y_train, 'ro')
plt.plot(x_test,predictions,'bo')
plt.xlabel('Features')
plt.ylabel('Score')
ax.grid()
plt.show()




```
